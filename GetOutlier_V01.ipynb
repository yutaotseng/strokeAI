```python
import numpy as np
import os
import shutil
from tqdm import tqdm
```

```python
# Set folder path
folder_path = 'waveform_like_data/no_stroke/normal_ecg'
dest_folder = 'waveform_like_data_without_outliers/no_stroke/normal_ecg'

# Create destination folder if it does not exist
os.makedirs(dest_folder, exist_ok=True)

# Loop through all the SVG files with tqdm progress bar
counter = 0
counter_outliers = 0
for filename in tqdm(os.listdir(folder_path), desc="Processing files"):
    file_path = os.path.join(folder_path, filename)
    try:
        data = np.load(file_path)
        lead, rhythm = data['arr_0'], data['arr_1']
        if rhythm.size != 0:
            if (lead > 450).sum() == 0 and (rhythm > 450).sum() == 0 and (lead < -450).sum() == 0 and (rhythm < -450).sum() == 0:
                shutil.copy(file_path, dest_folder)
                print(f"Copied {file_path} to {dest_folder}")
                counter += 1
            else:
                print("outlier:", filename)
                counter_outliers += 1
    except Exception as e:
        print(f"Error processing file {filename}: {e}")
        continue

print("good to go count:", counter)
print("outlier count:", counter_outliers)
print("-----DONE----")
```
